```python
import bs4 as bs
import urllib.request
import requests
import pandas as pd

from urllib.request import Request
from urllib.request import urlopen
import urllib
import random

# Set Random IP
def getHtml(url, proxies):
    proxy = random.choice(proxies)
    proxy_support = urllib.request.ProxyHandler({"http": proxy})
    opener = urllib.request.build_opener(proxy_support)
    urllib.request.install_opener(opener)
    html = urlopen(url)
    return html

proxies = ["101.53.101.172:9999", "101.53.101.172:9999", "171.117.93.229:8118", "119.251.60.37:21387",
           "58.246.194.70:8080"]

ratings = []
reviews = []
date = []
# reviewer = []
location = []
position =[]

import time
import random
time.sleep(random.random()*3)

def getHTMLText(url):
    maxTryNum = 20
    for tries in range(maxTryNum):
        try:
            kv = {"user-agent": "Mizilla/5.0"}
            response = requests.get(url, headers=kv, timeout=10)
            return response.text
        except:
            if tries < (maxTryNum - 1):
                continue
            else:
                print("Has tried %d times to access url %s, all failed!" % (maxTryNum, url))
                break

# n is the number of pages you wanna crawl, not bigger than the total page for specific company
n = 10
for i in range(0, 10):
    if i == 0:
        url = 'https://www.indeed.com/cmp/Bb&T/reviews?fcountry=ALL&lang=en'
    else:
        url = 'https://www.indeed.com/cmp/Bb&T/reviews?fcountry=ALL&start='+ str(20*i) + '&lang=en'

    page = requests.get(url)
    print(url)
    if page.status_code == requests.codes.ok:
        grURL = getHtml(url, proxies)
        grHTML = grURL.read()
        grURL.close()
        soup = bs.BeautifulSoup(grHTML, 'html.parser')
        rating_elements = soup.find_all("div", class_="cmp-ratingNumber")
        review_elements = soup.find_all("span", class_="cmp-review-text")
        date_elements = soup.find_all("span", class_="cmp-review-date-created")
        # reviewer_elements = soup.find_all("span", class_="cmp-reviewer")
        location_elements = soup.find_all("span", class_="cmp-reviewer-job-location")
        position_elements = soup.find_all("span", class_="cmp-reviewer-job-title")
        if not len(review_elements)==len(rating_elements)==len(date_elements)==len(location_elements)==len(position_elements):   #len(reviewer_elements)==
            continue
        for e in rating_elements:
            ratings.append(e.text)
        for e in review_elements:
            reviews.append(e.text)
        for e in date_elements:
            date.append(e.text)
        # for e in reviewer_elements:
        #     reviewer.append(e.text)
        for e in location_elements:
            location.append(e.text)
        for e in position_elements:
            position.append(e.text)
        print('page:',i,'\n') #'  reviewers:', len(reviewer),
df = pd.DataFrame({
            # 'Reviewer':reviewer,
            'Location':location,
            'Position': position,
            'Reviews': reviews,
            'Ratings': ratings,
            'Date':date})
file_path = 'E:\\GSU-Jingyi_Wang\\MSA8040-Data_Management\\Final_Project\\BB&T_withjobtitle.csv'
df.to_csv(file_path,columns=['Location','Position', 'Reviews', 'Ratings','Date'],index=False,sep=',') #'Reviewer',
```
